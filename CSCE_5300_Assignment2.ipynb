### 2D Plots

##### Line Plot: Create a line plot of sin(x) and cos(x + π/2) for -2π < x < 2π where x increases at intervals of π/4.

* Make the sin(x) graph red and make the cos(x+π/2) graph green
* Put both lines onto the same plot
* Using the same info as above, make a subplot with 2 different graphs one graph for sin(x) and one graph for cos(x+π/2)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline  
```

```python
x = np.arange(-2*np.pi,2*np.pi,np.pi/4)
y1 = np.sin(x)
y2 = np.cos(x+(np.pi/2))
```

```python
plt.plot(x,y1,'r',x,y2,'g')
plt.annotate('Sin(x) Function', xy=(0.7,0.6), xytext=(8, -0.75),arrowprops=dict(facecolor='black',width=0.5, headwidth=7),)
plt.annotate('Cos(x+pi/2) Function', xy=(0.7,-0.6), xytext=(8, 0.05),arrowprops=dict(facecolor='black',width=0.5, headwidth=7),)
plt.xlabel('x values from -2pi to 2pi (radians)')
plt.ylabel('sin(x) and cos(x+pi/2) (radians)')
plt.title('Plot of sin(x) and cos(x+pi/2) from -2pi to 2pi')
plt.legend(['sin(x)','cos(x+pi/2)'],bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.text(0,-1.65,"Caption: Both lines onto the same plot",ha="center",weight="bold")
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y1,'r')
plt.annotate('Sin(x) Function', xy=(-0.5,-0.5), xytext=(8, -0.75),arrowprops=dict(facecolor='black',width=0.5, headwidth=7),)
plt.xlabel('x values from -2pi to 2pi (radians)')
plt.ylabel('sin(x) (radians)')
plt.title('Plot of sin(x) from -2pi to 2pi')
plt.legend(['sin(x)'],bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.subplot(2,1,2)
plt.plot(x,y2,'g')
plt.annotate('Cos(x+pi/2) Function', xy=(0.7,-0.6), xytext=(8, -0.5),arrowprops=dict(facecolor='black',width=0.5, headwidth=7),)
plt.xlabel('x values from -2pi to 2pi (radians)')
plt.ylabel('cos(x+pi/2) (radians)')
plt.title('Plot of cos(x+pi/2) from -2pi to 2pi')
plt.legend(['cos(x+pi/2)'],bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.text(0,-2.65,"Caption: Both lines onto the different plots",ha="center",weight="bold")
plt.tight_layout()
plt.show()
```

### 3D Plots

##### Program for contour and surface (with heat maps) plots using the following trigonometric equation.

* Z= np.sin(x**2)+np.cos(y**2)

Both x and y values are in the interval [-pi, +pi] with delta=0.1. These points are an array of values starting from -pi to +pi in interval increments of 0.1.

```python
def f(x, y):
    return np.sin(x**2)+np.cos(y**2)

x = y = np.arange(-np.pi, np.pi, 0.1)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.figure(figsize=(10,6))
plt.contour(X, Y, Z, cmap='magma')
plt.xlabel('x values ranging from [-pi,pi] (radians)')
plt.ylabel('y values ranging from [-pi,pi] (radians)')
plt.title("Contour Plot")
plt.colorbar(label="Label for Color Codes in the Plot")
plt.text(0,-3.95,"Caption: 2D Cotour Plot for sin(x^2)+cos(y^2)",ha="center",weight="bold")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15,10))
ax = plt.axes(projection='3d')
cb = ax.contour3D(X, Y, Z, 50, cmap='hot')
plt.title("Surface Plot")
plt.colorbar(cb,label="Label for Color Codes in the Plot")
ax.set_xlabel('x values ranging from [-pi,pi] (radians)')
ax.set_ylabel('y values ranging from [-pi,pi] (radians)')
ax.set_zlabel('sin(x^2)+cos(y^2) (radians)')
ax.text(3,-3,-3,s='Caption: 3D Surface Plot for sin(x^2)+cos(y^2)',ha="center",weight="bold");
plt.show()
```
