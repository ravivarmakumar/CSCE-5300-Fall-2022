```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sports = ['Basketball','Baseball/Softball','Soccer','Track','Swimming','Tennis','Cross Country']
male = [40,24,20,17,9,8,5]
female = [25,17,17,17,12,8,4]
data = list(zip(sports,male,female))

dataframe = pd.DataFrame(data, columns = ['Sports','Male','Female'])
dataframe
```

```python
dataframe.plot(x="Sports", y=["Male", "Female"], kind="bar",figsize=(10,6),title="Side-by-Side Bar Chart for Male & Female",ylabel='Percentage(%)')
plt.tight_layout()
plt.show()
```

```python
dataframe.plot(x="Sports", y=["Male", "Female"], kind="bar",figsize=(10,6),title="Stacked Bar Chart for Male & Female",ylabel='Percentage(%)',stacked=True)
plt.tight_layout()
plt.show()
```
