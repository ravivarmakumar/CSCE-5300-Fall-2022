#### 4. Write a function EvenNums that takes a list of numbers [a,b].  Test the function on [1,100] and print the output. 

```python
def EvenNums(list):
    res = []
    a = list[0]
    b = list[1]
    for i in range(a,b+1):
        if i%2 == 0:
            res.append(i)
    return res
```

```python
print(EvenNums([2,100]))
```

### Part 2 (2 pts): Create PySpark DataFrame using Titanic Dataset “train.csv”. 

#### Repeat questions 1 through 3 from Part 1.  

```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Assignment3').getOrCreate()
spark
```

```python
sparkdataframe = spark.read.options(inferSchema=True,header=True,delimiter=',').csv('/Users/ravivarmakumarbevara/Desktop/Ph.D./CSCE_5300/train.csv')
sparkdataframe.show()
```

```python
from pyspark.sql.types import StringType
sparkdataframe = sparkdataframe.withColumn("Age",sparkdataframe.Age.cast(StringType()))
```

```python
sparkdataframe = sparkdataframe.fillna("",subset=None)
sparkdataframe.show()
```

```python
sparkdataframe.filter(sparkdataframe['Survived'] == 1).agg({"Age": "avg"}).show()
```

```python
sc  = spark.sparkContext

def EvenNumsPySpark(list):
    rdd = sc.parallelize(list)
    res = sc.parallelize([])
    a = rdd.first()
    b = rdd.take(2)[1]
    givenList = sc.range(a,b+1)
    res = givenList.filter(lambda x: x%2 == 0).collect()
    return res
```

```python
print(EvenNumsPySpark([2,100]))
```
