```python
import pandas as pd

data = [
 ["Children", "First", 6, 0],        
 ["Children", "Second", 24, 0],
 ["Children", "Third", 27, 52],
 ["Men", "First", 57, 118],
 ["Men", "Second", 14, 154],
 ["Men", "Third", 75, 387],
 ["Men", "Crew", 192, 693],
 ["Women", "First", 140, 4],
 ["Women", "Second", 80, 13],
 ["Women", "Third", 76, 89],
 ["Women", "Crew", 20, 3]
] 
 
pandasdataframe = pd.DataFrame(data, columns = ["Sex", "Class", "Survived", "Died"]) 
pandasdataframe
```

```python
import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Exam1').getOrCreate()
spark
```

```python
df_titanic = spark.createDataFrame(pandasdataframe) 
df_titanic.show()
```

```python
df_titanic_nofc = df_titanic.where(df_titanic.Class!='First')
df_titanic_nofc.show()
```

```python
df_titanic_updated = df_titanic.withColumn('Percent Died', df_titanic.Died/(df_titanic.Survived+df_titanic.Died))
df_titanic_updated.show()
```

```python
updatedpandasdataframe = df_titanic_updated.toPandas()
updatedpandasdataframe
```

```python
updatedpandasdataframe['Percent Died'] = updatedpandasdataframe['Percent Died'].round(2)
updatedpandasdataframe
```
