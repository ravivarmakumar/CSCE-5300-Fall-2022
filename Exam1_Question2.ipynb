```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    return np.exp(-2 * (X**2/4 + Y**2/4)) * np.sin(2 * np.pi * 0.2 * X)
```

```python
x = y = np.linspace(-6,6,100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.figure(figsize=(10,6))
plt.contour(X, Y, Z, cmap='magma')
plt.xlabel('Distances from Center in X-axis direction (cm)')
plt.ylabel('Distances from Center in Y-axis direction (cm)')
plt.title("Contour Plot",weight='bold')
plt.colorbar(label="Label for Color Codes in the Plot")
plt.text(0,-8,"Caption: Cotour Plot for visual response properties of a vertically oriented simple cell receptive field in the primary visual cortex",ha="center",weight="bold")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.contourf(X, Y, Z, cmap='gray')
plt.xlabel('Distances from Center in X-axis direction (cm)')
plt.ylabel('Distances from Center in Y-axis direction (cm)')
plt.title("Contour Plot for Positve and Negative Regions",weight='bold')
plt.colorbar(label="Label for degree of darker and lighther regions")
plt.text(0,-8,"Caption: B/W Cotour Plot for visual response properties of a vertically oriented simple cell receptive field in the primary visual cortex",ha="center",weight="bold")
plt.tight_layout()
plt.show()    
```
